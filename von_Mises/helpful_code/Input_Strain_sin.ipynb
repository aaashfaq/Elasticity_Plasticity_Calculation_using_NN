```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate strain history
def generate_strain_history(epsilon_0, frequency, num_samples, f_t_case="constant", a=1):
    # Time vector
    t = np.linspace(0, 1, num_samples)  # 1 second duration
    # Sine wave base
    base_sine = epsilon_0 * np.sin(2 * np.pi * frequency * t)
    
    # Case 1: f(t) = 1 (constant)
    if f_t_case == "constant":
        f_t = 1
        strain_history = base_sine * f_t
    
    # Case 2: f(t) = a * t (linearly increasing function)
    elif f_t_case == "linear":
        f_t = a * t
        strain_history = base_sine * f_t
    
    else:
        raise ValueError("Invalid case for f(t). Choose 'constant' or 'linear'.")
    
    return t, strain_history

# Parameters
epsilon_0 = 0.1  # Amplitude of strain
frequency = 1    # Frequency in Hz
num_samples = 1000  # Number of time samples
a = 2  # Scaling factor for linear case

# Generate strain history for both cases
t_constant, strain_constant = generate_strain_history(epsilon_0, frequency, num_samples, f_t_case="constant")
t_linear, strain_linear = generate_strain_history(epsilon_0, frequency, num_samples, f_t_case="linear", a=a)

# Plot the results
plt.figure(figsize=(10, 6))

# Plot for f(t) = 1
plt.subplot(2, 1, 1)
plt.plot(t_constant, strain_constant, label=r"$f(t) = 1$")
plt.title("Strain History: $f(t) = 1$")
plt.xlabel("Time (s)")
plt.ylabel("Strain")
plt.grid(True)

# Plot for f(t) = a * t
plt.subplot(2, 1, 2)
plt.plot(t_linear, strain_linear, label=r"$f(t) = at$", color="orange")
plt.title("Strain History: $f(t) = at$")
plt.xlabel("Time (s)")
plt.ylabel("Strain")
plt.grid(True)

plt.tight_layout()
plt.show()
```
